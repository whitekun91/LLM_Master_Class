````python
def customer_handler(name, ranking, language="ko"):
  # 월클 VIP 관!
  greetings =  {
      "ko": {"VIP": f"어서 오십시오 우리의 매출 실적을 구해주실 호갱 {name}님!!",
             "normal": f"안녕하세요! {name}씨"},
      "fr": {"VIP": f"Tu vas nous donner beaucoup d'argent!! Nous t'aimons {name}!!"},
             "normal": f"Salut! {name}, une personne de valeur base"}
  if ranking == "VIP":
    print(greetings[language]["VIP"])
  else:
    print(greetings[language]["normal"])
````

````python
def simple_calculator(operation, a, b):
    """
    기본 산술 연산을 수행하는 간단한 계산기 함수입니다.

    인수:
    operation (str): 수행할 연산. 'add', 'subtract', 'multiply', 또는 'divide'가 될 수 있습니다.
    a (float): 첫 번째 숫자.
    b (float): 두 번째 숫자.

    반환값:
    float 또는 str: 산술 연산의 결과 또는 오류 메시지.
    """
    if operation == 'add':
        return a + b
    elif operation == 'subtract':
        return a - b
    elif operation == 'multiply':
        return a * b
    elif operation == 'divide':
        if b == 0:
            return "오류: 0으로 나누기는 허용되지 않습니다."
        else:
            return a / b
    else:
        return "잘못된 연산입니다. 'add', 'subtract', 'multiply', 'divide' 중에서 선택해주세요."
````

````python
# 사용 예:
simple_calculator('add', 5, 3)      # 8을 반환합니다
````

````python
simple_calculator('subtract', 5, 3) # 2를 반환합니다
````

````python
simple_calculator('multiply', 5, 3) # 15를 반환합니다
````

````python
simple_calculator('divide', 5, 3)   # 대략 1.67을 반환합니다
````

````python
simple_calculator('divide', 5, 0)   # "오류: 0으로 나누기는 허용되지 않습니다."를 반환합니다
````

````python
simple_calculator('hello', 5, 3)  # "잘못된 연산입니다."를 반환합니다
````

에러 체킹도 넣고 함수화 한 버전으로 봅시다

````python

import requests

def get_cat_fact():
    url = 'https://catfact.ninja/fact'
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raises an HTTPError if the response status code is 4XX or 5XX
        cat_fact = response.json()  # Converts the response to JSON
        return cat_fact['fact']
    except requests.RequestException as e:  # Handles exceptions
        print(f"An error occurred: {e}")
        return None

````

````python
# Example usage
cat_fact = get_cat_fact()
if cat_fact:
    print(f"냥이 팩트입니다: {cat_fact}")
else:
    print("에러 났습니다.")
````

# 배달 앱 예

1. UI 로 주문을 받아서
2. 그 주문을 API 콜에 맞는 주문서로 작성을 하여
3. 주문서를 HTTP call 로 접수함.

API 콜: 주문서를 주문 받는 사람에게 보냄
HTTP call: 보내는 방법 (전화, 팩스, 우편, 직접 방문...)
REST API: 주문서 보내는 여러 방법 중 업계 표준 (전화!!)

현재 거의 모든 앱/프로그래밍의 구조:

- 백엔드 (실제 일처리를 함)
- 프론트엔드 (유저들이 백엔드를 쓸 수 있도록 함. 인터넷 브라우저. 모바일 앱 등등)

- 프론트엔드에서 백엔드를 연결하는 방법:  
  - 백만년 전: 메인 프레임 + 터미널 등등
  - 인터넷 도입 후: 웹사이트 + 웹서버 + 백엔드 서버 등등
  - 모바일: 모바일 앱 + 백엔드 서버

백엔드 부르는 방법: 아주 여러가지 있지만 하나만 알면 됨 - REST API



- data formats: csv, tsv, different separators
- json format
